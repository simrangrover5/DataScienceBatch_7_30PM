```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import warnings
```

```python
warnings.filterwarnings(action="ignore")
```

```python
plt.rcParams['figure.figsize'] = 11, 5
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.labelcolor'] = 'red'
plt.rcParams['axes.labelsize'] = 10
```

```python
tip = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
car = sns.load_dataset("car_crashes")
flight = sns.load_dataset("flights")
```

```python
sns.violinplot(tip['tip'])
plt.show()
```

```python
# map{"male": "b", "female": "y"}
```

```python
sns.violinplot(x="day", y="total_bill", data=tip, hue="sex", split=True, palette={"Male": "b", "Female": "w"},
              inner="quart")
plt.show()
```

```python
exer = sns.load_dataset("exercise", index_col=0)
```

```python
exer.head()
```

```python
sns.catplot(x="time", y="pulse", data=exer, col="diet", kind="point")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.factorplot(x="time", y="pulse", data=exer, col="diet", hue="kind")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-5, 5, 20)
```

```python
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = 11, 5
```

```python
plt.plot(x, x**2)
```

```python
lines = plt.plot(x, x**2, 'ro--', x, x**3, 'g^-.')
plt.legend(lines, ["$X^2$", "$X^3$"])
plt.show()
```

```python
fig = plt.figure(facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(x, x**2, 'ro--', label="$X^2$")
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
ax.spines['bottom'].set_color("grey")
ax.spines['left'].set_color("grey")
plt.tick_params(color="green", size=10, labelcolor="coral", labelsize=15, width=10)
plt.title("X Vs $X^2$", fontdict=title)
plt.legend(frameon=False, labelcolor="white")
plt.grid(alpha=0.5, lw=0.8, ls='--', color='#123456', which="major", axis='y')
plt.show()
```

```python
fig = plt.figure(facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(x, x**2, 'ro--', label="$X^2$")
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
ax.spines['bottom'].set_color("grey")
ax.spines['left'].set_color("grey")
plt.tick_params(color="green", size=10, labelcolor="coral", labelsize=15, width=10)
plt.title("X Vs $X^2$", fontdict=title)
plt.legend(frameon=False, labelcolor="white")
plt.grid(alpha=0.5, lw=0.8, ls='--', color='#123456', which="major", axis='x')
plt.show()
```

```python
fig = plt.figure(facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(x, x**2, 'ro--', label="$X^2$")
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
ax.spines['bottom'].set_color("grey")
ax.spines['left'].set_color("grey")
plt.tick_params(color="green", size=10, labelcolor="coral", labelsize=15, width=10)
plt.title("X Vs $X^2$", fontdict=title)
plt.legend(frameon=False, labelcolor="white")
plt.grid(alpha=0.5, lw=0.8, ls='--', color='#123456', which="major")
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
plt.scatter(iris['petal_length'])
```

```python
plt.scatter(iris['petal_length'], iris['petal_width'], color='red')
plt.xticks(np.arange(1, 8, 0.5))
plt.show()
```

```python
iris['species']
```

```python
color = iris['species'].map({'setosa': 1, 'versicolor': 2, 'virginica': 3})
```

```python
plt.scatter(iris['petal_length'], iris['petal_width'], c=iris['species'])
plt.xticks(np.arange(1, 8, 0.5))
plt.show()
```

```python
plt.scatter(iris['petal_length'], iris['petal_width'], c=color)
plt.xticks(np.arange(1, 8, 0.5))
plt.show()
```

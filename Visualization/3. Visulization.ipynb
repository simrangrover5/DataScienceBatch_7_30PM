```python
data['day'].unique()
```

```python
size = data['day'].map({"Fri":20, "Sat":30, "Sun":40, "Thur":50})
color = data['time'].map({"Dinner":'#abcdef', 'Lunch':'coral'})
```

```python
plt.scatter(data['total_bill'], data['tip'], marker='*', s=size, c=color)
#plt.axis("off")
plt.tick_params(labelcolor="white", labelsize=12)
```

```python
data.head()
```

```python
size = data['smoker'].map({"Yes":20, "No":50})
plt.scatter(data['total_bill'], data['tip'], marker='*', s=size, c=color)
#plt.axis("off")
plt.tick_params(labelcolor="white", labelsize=12)
plt.xticks(range(0, 51, 5))
plt.plot([], [], label="Dinner", color="#abcdef")
plt.plot([], [], label="Lunch", color="coral")
plt.plot([], [], label="SMOKERS-20")
plt.plot([], [], label="NON SMOKERS-50")
plt.xlabel("TOTAL_BILL")
plt.ylabel("TIP")
plt.legend()
plt.grid(alpha=0)
```

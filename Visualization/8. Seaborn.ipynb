```python
sns.countplot(mat['winner'], hue=mat['toss_decision'], alpha=0.5, dodge=False)
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(facecolor="black")
ax1 = fig.add_axes([0, 0, 0.6, 0.6], facecolor="black")
ax2 = fig.add_axes([0.7, 0.7, 0.6, 0.6], facecolor="black")
sns.countplot(mat['winner'], ax=ax2)
sns.countplot(mat['winner'], hue=mat['toss_decision'], alpha=0.5, dodge=False, ax=ax1)
for a in [ax1, ax2]:
    a.tick_params(labelcolor="white")
    a.set_xticklabels(a.get_xticklabels(), rotation=90)
ax1.set_title("WINNER COUNT", color="coral")
ax2.set_title("TOSS DECISION BY TEAMS", color="coral")
plt.show()
```

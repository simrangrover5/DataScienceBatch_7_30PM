```python
dates = pd.date_range("160121", periods=10)
jaipur = np.random.randint(50, 100, 10)
delhi = np.random.randint(50, 100, 10)
```

```python
plt.bar(dates, jaipur, label="JAIPUR", alpha=0.5)
plt.bar(dates, delhi, label="DELHI", alpha=0.5)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
rad = np.linspace(0, 2*np.pi, 30)
sin = np.sin(rad)
cos = np.cos(rad)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3)
ax1.plot(rad, sin, 'r--')
ax2.plot(rad, cos, 'm*-.')
ax3.plot(rad, np.tan(rad), 'gx-')
ax1.stem(rad, sin, '--')
ax2.stem(rad, cos, '-.')
ax3.stem(rad, np.tan(rad), '--')
plt.show()
```

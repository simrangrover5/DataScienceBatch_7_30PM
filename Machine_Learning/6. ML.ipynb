```python
height = np.random.normal(200, 50, 200)
weight = 0.3*height + np.random.normal(5, 2, 200)
plt.plot(height, weight, 'o')
plt.xlabel("HEIGHT")
plt.ylabel("WEIGHT")
```

$$ y = mx + c $$

    y --> target
    x --> feature
    
    
    x --> height
    y --> weight

```python
y = weight.copy()
x = height.copy()
y = mx + c
```

### There are 2 ways to find the value of m and c or 
$\beta_0$ and $\beta_1$


    1. Hit and trial method
    2. Formula

```python
m = 0.1
c = 0
x = height.copy()
y = m*x + c  # learning 
plt.plot(height, weight, 'o')
plt.plot(x, y, '--')
```

```python
m = 1
c = 0
x = height.copy()
y = m*x + c  # learning 
plt.plot(height, weight, 'o')
plt.plot(x, y, '--')
```

```python
m = 0.33
c = 0.05
x = height.copy()
y = m*x + c  # learning 
plt.plot(height, weight, 'o')
plt.plot(x, y, 'r-')
```

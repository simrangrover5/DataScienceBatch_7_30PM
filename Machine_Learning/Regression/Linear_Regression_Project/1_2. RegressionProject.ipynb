```python
features = data.drop("median_house_value", axis=1)
target = data['median_house_value']
```

```python
features.columns[:-1]
```

```python
features.columns[-1]
```

```python
num_attr = features.columns[:-1]
cat_attr = features.columns[-1]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
num_pipe = Pipeline([
    ('simple', SimpleImputer())
])
cat = OneHotEncoder()
```

```python
f1 = num_pipe.fit_transform(features[num_attr])
```

```python
f2 = cat.fit_transform(features[['ocean_proximity']]).toarray()
```

```python
cat.categories_
```

```python
f1.shape
```

```python
f2.shape
```

```python
np.hstack([f1, f2])
```

```python
final = np.hstack([f1, f2])
columns = np.append(features.columns[:-1], cat.categories_)
df_feat = pd.DataFrame(final, columns=columns)
```

```python
df_feat
```

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attr
```

```python
cat_attr = [cat_attr]
```

```python
pipe = ColumnTransformer([
    ('num', num_pipe, num_attr),
    ('cat', OneHotEncoder(), cat_attr)
])
```

```python
final_feat = pipe.fit_transform(features)
```

```python
final.shape
```

```python
final_feat.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(final_feat, target, random_state=100)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
pred = model.predict(X_test)
mean_absolute_error(Y_test, pred)
```

```python
r2_score(Y_test, pred)
```

```python
num_pipe = Pipeline([
    ('simple', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])
```

```python
pipe = ColumnTransformer([
    ('num', num_pipe, num_attr),
    ('cat', OneHotEncoder(), cat_attr)
])
final_feat = pipe.fit_transform(features)

X_train, X_test, Y_train, Y_test = train_test_split(final_feat, target, random_state=10)


model = LinearRegression()
model.fit(X_train, Y_train)
pred = model.predict(X_test)
mean_absolute_error(Y_test, pred)

```

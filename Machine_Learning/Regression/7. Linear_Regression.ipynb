$$ error = \frac {\sum\limits_{i=0}^N (y_i - \hat y_i)} {N} $$

```python
error = 0
for yi, y_hat in zip(y, y1):
    error += (yi - y_hat)
print(error/len(y))
```

$$ \text{mean absolute error(mae)} = \frac {\sum\limits_{i=0}^N | (y_i - \hat y_i) |}{N} $$

```python
def error(actual, pred):
    return (sum(np.abs(actual-pred))/len(actual))
```

```python
# actual x --> 136.71
# actual y --> 273
# perdicted y --> 320.77
m = 2.2
c = 20
y1 = m*x + c  # x --> training features, y --> actual target, y1 --> predicted target
error(y, y1)
```

```python
for m in np.arange(2, 3.5, 0.05):
    for c in np.arange(1, 51, 10):
        pred = m*x + c
        plt.plot(x, y, 'o', label='ACTUAL')
        plt.plot(x, pred, 'r--', label="PREDICTED")
        err = error(y, pred)
        plt.title(f"{m} and {c} and {err}")
        plt.show()
```

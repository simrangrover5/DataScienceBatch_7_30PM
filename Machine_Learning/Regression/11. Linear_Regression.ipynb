```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.random.normal(10, 7, 100)
y = 2*x + np.random.normal(5, 3, 100)
plt.plot(x, y, 'o')
```

```python
x = np.random.normal(10, 7, 100)
y = 2*x + np.random.normal(5, 3, 100)
```

```python
m = 1.9
c = 0.1
y1 = m*x + c
plt.plot(x, y, 'o')
plt.plot(x, y1, '-')
plt.show()
```

```python
abs(y - y1).sum()/len(y)  # bias 
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, y1)
```

```python
test_x = np.random.normal(10, 7, 50)
test_y = 2*test_x + np.random.normal(5, 3, 50)
plt.plot(test_x, test_y, 'o')
```

```python
m = 1.9
c = 0.1
pred = m*test_x + c
plt.plot(test_x, test_y, 'o')
plt.plot(test_x, pred, '--')
r2_score(test_y, pred)  # 100%
```

```python
abs(test_y - pred).sum()/len(pred)  # variance
```

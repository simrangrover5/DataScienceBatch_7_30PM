```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = np.random.randint(10, 20, 20)
```

```python
X = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

$$ y = ax^2 + bx + c $$

```python
m = 0.3
c = 1
Y = []
for x_i in X:
    Y.append(m*x_i + c)
print(Y)
```

```python
plt.plot(X, Y)
```

```python
def deviation(arr):
    return abs(arr - arr.mean()).sum()/len(arr)
```

```python
plt.figure(dpi=150, facecolor="grey")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("grey")
plt.plot(ages, 'w*')
m = ages.mean()
dev = deviation(ages)
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.plot([0, 20], [m + dev, m+dev], 'w--')
plt.plot([0, 20], [m-dev, m-dev], 'm--')
plt.xticks(range(20))
plt.tick_params(labelcolor="black")
plt.legend()
plt.show()
```

```python
dev
```

### Variance

$$ variance = \frac {\sum\limits_{i=0}^N (x_i - \bar x)^2} {N} $$

```python
def variance(arr):
    return ((arr - arr.mean())**2).sum()/len(arr)
```

```python
variance(ages)
```

```python
dev
```

```python
plt.figure(dpi=150, facecolor="#123456")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("#123456")
plt.plot(ages, 'w*')
m = ages.mean()
dev = deviation(ages)
var = variance(ages)
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.plot([0, 20], [m + dev, m+dev], 'w--')
plt.plot([0, 20], [m-dev, m-dev], 'w--')
plt.plot([0, 20], [m + var, m+var], 'm--')
plt.plot([0, 20], [m-var, m-var], 'm--')
plt.xticks(range(20))
plt.tick_params(labelcolor="black")
plt.show()
```

### Standard Deviation

$$ Std = \sqrt{\frac {\sum\limits_{i=0}^N (x_i - \bar x)^2} {N}} $$

```python
np.sqrt(var)
```

```python
dev
```

```python
plt.figure(dpi=150, facecolor="#123456")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("#123456")
plt.plot(ages, 'w*')
m = ages.mean()
dev = deviation(ages)
var = variance(ages)
std = np.sqrt(var)
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.plot([0, 20], [m + dev, m+dev], 'w--')
plt.plot([0, 20], [m-dev, m-dev], 'w--')
plt.plot([0, 20], [m + var, m+var], 'm--')
plt.plot([0, 20], [m-var, m-var], 'm--')
plt.plot([0, 20], [m + std, m+std], 'g--')
plt.plot([0, 20], [m-std, m-std], 'g--')
plt.xticks(range(20))
plt.tick_params(labelcolor="black")
plt.show()
```

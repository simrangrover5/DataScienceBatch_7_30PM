### Merge

    To merge the dataframes or to join the dataframe such that you can use both the dataframe 
    data simultanously
    
        1. inner
        2. outer
        3. left
        4. right

```python
df1 = pd.DataFrame({
    "Id" : range(1, 11),
    "Name" : [chr(i) for i in range(65, 75)],
    "Maths" : np.random.randint(1, 50, 10)
})

df2 = pd.DataFrame({
    "Id" : range(1, 11),
    "Science" : np.random.randint(1, 50, 10),
    "English" : np.random.randint(1, 50, 10),
})
```

```python
df1
```

```python
df2
```

```python
### Name of students having marks in eng > 30
```

```python
df1[df2['English']>30]['Name']
```

    inner join
        select * from table1 inner join table2 on table1.col = table2.col;
    

```python
pd.merge(df1, df2)
```

```python
df2["Id"] = list(range(1, 10, 2)) + list(range(11, 16))
```

```python
df2
```

```python
pd.merge(df1, df2)  # default = inner join
```

```python
df3 = df2.copy()
df3.rename(columns={"Id" : "Sid"}, inplace=True)
```

```python
df3
```

```python
pd.merge(df1, df3, left_on="Id", right_on="Sid")
```

```python
pd.merge(df1, df2, on="Id")
```

```python
help(pd.merge)
```

```python
pd.merge(df1, df2, left_index=False)
```

```python
df3.index = range(11, 21)
```

```python
df3
```

```python
df1.merge(df3, left_on="Id", right_on="Sid")
```

```python
df3.merge(df1, left_on="Sid", right_on="Id", left_index=True)
```

```python
df3.merge(df1, left_on="Sid", right_on="Id", right_index=True)
```

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df1, df2, how="inner", left_on="Maths", right_on="Science")
```

```python
df2
```

```python
pd.merge(df1, df2, how="outer", on="Id")
```

```python
pd.merge(df1, df2, how="outer", left_on="Maths", right_on="Science")
```

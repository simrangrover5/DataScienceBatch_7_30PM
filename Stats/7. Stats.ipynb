## Median

    1. It depends on number of values not on the values
    2. Thats why it does not get effected by outliers
    
        For this the data should be sorted

#### median for odd number of values

$$ i = \frac {N+1} {2} $$
 
The array is arr and N is the number of terms

$$ median = arr[i] $$

```python
a1 = list(ages)
```

```python
def median(arr):
    arr.sort()
    N = len(arr)
    if N%2 != 0:
        i = (N+1)//2  # 21 + 1//2 --> 22//2 --> 11
        m = arr[i-1]  # 11 term means element at index 10
        return m
    else:
        print("EVEN WORK")
```

#### median for even number of values

$$ i_1 = \frac {N} {2} $$

$$ i_2 = \frac {N+2} {2} $$
 
The array is arr and N is the number of terms

$$ median = \frac {arr[i_1-1]  + arr[i2-1]} {2} $$

```python
def median(arr):
    arr.sort()
    N = len(arr)
    if N%2 != 0:
        i = (N+1)//2  # 21 + 1//2 --> 22//2 --> 11
        m = arr[i-1]  # 11 term means element at index 10
    else:
        i1 = (N)//2
        i2 = (N+2)//2
        m = (arr[i1-1] + arr[i2-1])/2
    return m
```

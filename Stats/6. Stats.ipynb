### Mean
     1. Calculate Average
     2. It is effected by large values or outliers
     3. It takes all oberservations into account
     4. It is not so meaningfull when there is single value
 
##### Formula 

$$ mean = \frac {\sum\limits_{i=0}^N x_i} {N} $$

$$ mean = \frac {x_0 + x_1 + x_2 + ......+ x_{n-1}} {N} $$

$$ mean = \frac {x_1 + x_2 + ......+ x_n} {N} $$

```python
def mean(arr):
    sum_ = 0
    for i in arr:
        sum_ += i
    m_ = sum_ / len(arr)
    return m_
```

```python
font = {"fontsize":20, "color":"red"}
m = mean(x)
plt.plot(x, 'o')
plt.plot([0, 20], [m, m], 'k--')  #[x1, x2], [y1, y2], k --> black -- --> linestyle
plt.xlabel("INDEX", fontdict=font)
plt.ylabel("X VALUE", fontdict=font)
plt.xticks(range(20))
plt.show()

# x1, y1 --> 0, 15.95(m)
# x2, y2 --> 20, 15.95(m)
```

```python
font = {"fontsize":20, "color":"red"}
ax = plt.gca()  # axes
m = mean(x)
plt.plot(x, 'o')
plt.plot([0, 20], [m, m], 'k--')  #[x1, x2], [y1, y2], k --> black -- --> linestyle
plt.xlabel("INDEX", fontdict=font)
plt.ylabel("X VALUE", fontdict=font)
plt.xticks(range(20))
plt.tick_params(labelsize=15, labelcolor="green")
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.show()

# x1, y1 --> 0, 15.95(m)
# x2, y2 --> 20, 15.95(m)
```

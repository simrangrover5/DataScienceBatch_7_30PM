```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.DataFrame({
    "Id" : range(1, 11),
    "Name" : [chr(i) for i in range(65, 75)],
    "Maths" : np.random.randint(1, 50, 10)
})

df2 = pd.DataFrame({
    "Id" : range(1, 11),
    "Science" : np.random.randint(1, 50, 10),
    "English" : np.random.randint(1, 50, 10),
})
```

```python
df1
```

```python
df2
```

```python
df3 = df2.copy()
df3.rename(columns={"Id" : "Sid"}, inplace=True)
```

```python
df3.index = list(range(3, 12, 2)) + [12, 13, 14, 15, 16]
```

```python
df3
```

```python
df3['Sid'] = list(range(3, 12, 2)) + [12, 13, 14, 15, 16]
```

```python
pd.concat((df1, df3), join="inner", axis=1)
# axis=0 --> columns
# axis=1 --> index
```

```python
pd.concat((df1, df3), join="outer", axis=0, ignore_index=True)
# axis=0 --> columns
# axis=1 --> index
```

```python
pd.concat((df1, df3), ignore_index=True)
# axis=0 --> columns
# axis=1 --> index
```

```python
pd.concat((df1, df3), ignore_index=False)
# axis=0 --> columns
# axis=1 --> index
```

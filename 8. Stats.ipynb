```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def median(arr):
    arr.sort()
    N = len(arr)
    if N%2 != 0:
        i = (N+1)//2  # 21 + 1//2 --> 22//2 --> 11
        m = arr[i-1]  # 11 term means element at index 10
    else:
        i1 = (N)//2
        i2 = (N+2)//2
        m = (arr[i1-1] + arr[i2-1])/2
    return m
```

```python
ages = np.random.randint(10, 20, 20)
```

```python
ages.mean()
```

```python
np.median(ages)
```

```python
plt.figure(dpi=150, facecolor="black")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(ages, 'w*')
m = ages.mean()
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
med = median(ages)
plt.plot([0, 20], [med, med], '--', label="MEDIAN")
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.xticks(range(20))
plt.tick_params(labelcolor="grey")
plt.legend()
plt.show()
```

```python
def mode(arr):
    count = {}
    for ele in arr:
        if ele in count:
            count[ele] += 1
        else:
            count[ele] = 1  # creating new key with the name as ele and initializing it with 1
    temp = sorted(count.items()) # sort with keys and store in temp
    ele = 0
    maxi = 0
    for key, value in temp:
        if maxi < value:
            maxi = value
            ele = key
    return ele
```

```python
class Stats:
    def __init__(self, arr):
        self.arr = arr
        self.N = len(arr)
        
    def mean(self):
        return sum(self.arr)/self.N

    def median(self):
        self.arr.sort()
        if self.N%2 != 0:
            i = (self.N+1)//2  # 21 + 1//2 --> 22//2 --> 11
            m = self.arr[i-1]  # 11 term means element at index 10
        else:
            i1 = (self.N)//2
            i2 = (self.N+2)//2
            m = (self.arr[i1-1] + self.arr[i2-1])/2
        return m

    def mode(self):
        count = {}
        for ele in self.arr:
            if ele in count:
                count[ele] += 1
            else:
                count[ele] = 1  # creating new key with the name as ele and initializing it with 1
        temp = sorted(count.items()) # sort with keys and store in temp
        ele = 0
        maxi = 0
        for key, value in temp:
            if maxi < value:
                maxi = value
                ele = key
        return ele
```

```python
class One:
    __slots__ = ['__name', 'age']
    def __init__(self, name, age=None):
        self.__name = name  # private variable inside class --> self.__name but outside class 
        self.age = age
        
class Two(One):
    __slots__ = ['course', "fees"]
    def __init__(self, name, course, fees):
        super().__init__(name)
        self.course = course
        self.fees = fees
    
    def show(self):
        print("\n Name : ", self.__name)
        print("\n Course : ", self.course)
        print("\n Fees : ", self.fees)
        
obj = Two("simran", "python", 12000)
obj.show()
```

```python
for i in soup.select("li"):
    print(i.text)
```

```python
for ele in soup.select("a"):
    print(ele.text)
```

```python
for ele in soup.select("a"):
    print(ele['href'])  # attribute --> ele[attribute_name]
```

```python
"hello".startswith("h")
```

```python
for ele in soup.select("a"):
    if ele['href'].startswith("www"):
        print("https://"+ele['href'])
    elif "https" not in ele["href"]:
        print("https://www.parsehub.com" + ele['href'])
    else:
        print(ele['href'])
```

```python
for i in soup.select("img"):
    print(i['src'])
```

```python
for i in soup.select("img"):
    if "https" not in i["src"]:
        print("https://www.parsehub.com" + i['src'])
    else:
        print(i['src'])
```

```python
url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY"
resp = requests.get(url)
```

```python
resp.status_code
```

```python
flip = BeautifulSoup(resp.text, 'html')
```

```python
flip.select("._4rR01T") # use . before class
```

```python
name = []
for t in flip.select("._4rR01T"):
    name.append(t.text)
```

```python
flip.findAll("div", attrs={"class":'_4rR01T'})
```

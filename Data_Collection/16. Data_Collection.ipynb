```python
def show_details(title):
    url = f"http://www.omdbapi.com/?apikey=461a20c0&t={title}"
    resp = requests.get(url)
    if resp.status_code == 200:
        data = resp.json()
        keys = ['Year', 'Plot', 'Rated', 'Language', 'Director', 'Actors']
        print("DETAILS".center(80, "*"))
        for k in keys:
            print(f"{k} ---> {data.get(k)}".center(80))
        if data.get("Ratings"):
            print(f"Rating ---> {data['Ratings'][0]['Value']}".center(80))
        try:
            with open(f"{title}.jpg", 'wb') as fp:
                img = requests.get(data['Poster'])
                fp.write(img.content)
                img = plt.imread(f"{title}.jpg")
                plt.figure(figsize=(11, 5))
                plt.imshow(img)
                plt.xticks([])
                plt.yticks([])
                plt.show()
        except:
            pass
        print("*"*80)
    else:
        print("\n SOMETHING GOT WRONG")
```

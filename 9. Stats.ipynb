```python
ages = np.random.randint(10, 20, 20)
```

```python
plt.figure(dpi=150, facecolor="black")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(ages, 'w*')
m = ages.mean()
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.xticks(range(20))
plt.tick_params(labelcolor="grey")
plt.legend()
plt.show()
```

#### Deviation
    
    Distance from mean
    How far the point is from mean
    
    
$$ dev = \frac {\sum\limits_{i=0}^N (x_i - \bar x)} {N}  $$

```python
ages - ages.mean()
```

```python
sum(ages - ages.mean())/len(ages)
# dev --> 0 
```

 $$ dev = \frac {\sum\limits_{i=0}^N |x_i - \bar x|} {N}  $$

```python
abs(ages - ages.mean())
```

```python
dev = abs(ages - ages.mean()).sum()/len(ages)
```

```python
dev
```

    spread --> mean + dev, mean - dev

```python
plt.figure(dpi=150, facecolor="black")  # dpi --> dots per inch
ax = plt.gca()
ax.set_facecolor("black")
plt.plot(ages, 'w*')
m = ages.mean()
plt.plot([0, 20], [m, m], 'r--', label="MEAN") # x1 --> 0, x2 --> 20, y1 --> mean , y2 --> mean
for i in range(len(ages)):
    plt.plot([i, i], [ages[i], m], '*-') 
plt.plot([0, 20], [m + dev, m+dev], 'w--')
plt.plot([0, 20], [m-dev, m-dev], 'm--')
plt.xticks(range(20))
plt.tick_params(labelcolor="grey")
plt.legend()
plt.show()
```

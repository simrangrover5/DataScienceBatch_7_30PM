```python
keys = ['deaths', 'cases', 'active_cases', 'total_recovered', 'new_cases', 'new_deaths', 'country_name']
deaths = []
cases = []
active_cases = []
total_recovered = []
new_cases = []
new_deaths = []
country_name = []
for value in data['countries_stat']:
    deaths.append((value.get("deaths").replace(",", '')))
    cases.append((value.get("cases").replace(",", '')))
    active_cases.append((value.get('active_cases').replace(",", '')))
    total_recovered.append((value.get('total_recovered').replace(",", '')))
    new_cases.append((value.get("new_cases").replace(",", '')))
    new_deaths.append((value.get("new_deaths").replace(",", '')))
    country_name.append((value.get("country_name")))
corona = pd.DataFrame({
    "Countries" : country_name,
    "Active_Cases" : active_cases,
    "Total_Recovered" : total_recovered,
    "Deaths" : deaths,
    "Cases" : cases,
    "New_Cases" : new_cases,
    "New_Deaths" : new_deaths
})
```
